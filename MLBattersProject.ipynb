```python
import pandas as pd
```

```python
# dataset of hall of famers from 1871-2016
# players only, no coaches
# players inducted after 1900 (common era)

hof = pd.read_csv("hall_of_fame.csv", index_col=0)
hof = hof[hof["category"] == "Player"]
hof = hof[hof["yearid"] >= 1901]
```

```python
# dataset of batters' seasons from 1871-2015
# batters from after 1900 (common era)

batters = pd.read_csv("batters.csv", index_col=0)
batters = batters[batters["year"] >= 1901]
```

```python
# dataset of batters' seasons with non-statistical values
# (i.e. year, stint, team, league)

batters_nonstats = batters.iloc[:, :4]
batters_nonstats
```

```python
# datasets of batters' seasons with ONLY statistical values
# (i.e. games, at-bats, runs, hits, rbi, etc.)

batters_stats = batters.iloc[:, 4:]
batters_stats
```

```python
# dataset of batters' season averages

batters_avg = batters_stats[batters_stats["g"] >= 100].groupby('player_id').mean()
batters_avg
```

```python
# dataset of hall of famers and potential hall of famers
# and their season average stats

hof_stats = pd.merge(hof, batters_avg, how="inner", on="player_id")
hof_stats
```
