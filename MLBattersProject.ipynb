```python
import pandas as pd
```

dataset of hall of famers from 1871-2016

```python
hof = pd.read_csv("hall_of_fame.csv", index_col=0)
```

players only, no managers

```python
hof = hof[hof["category"] == "Player"]
```

players from after 1900

```python
hof = hof[hof["yearid"] >= 1901]
```

dataset of batters' seasons from 1871-2015

```python
batters = pd.read_csv("batters.csv", index_col=0)
```

batters from after 1900



```python
batters = batters[batters["year"] >= 1901]
```

dataset of batters' seasons with non-stats

```python
batters_nonstats = batters.iloc[:, :4]
```

```python
batters_nonstats
```

dataset of batters' seasons and only stats

```python
batters_stats = batters.iloc[:, 4:]
```

```python
batters_stats
```

dataset of a batters' season averages of seasons with more than 100 games played

```python
batters_avg = batters_stats[batters_stats["g"] >= 100].groupby('player_id').mean()
```

```python
batters_avg
```

dataset of hall of famers and potential hall of famers and their season average stats

```python
hof_stats = pd.merge(hof, batters_avg, how="inner", on="player_id")
```
